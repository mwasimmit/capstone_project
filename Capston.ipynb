```python
def graana(url):
    types=[] #List of type of property
    prices=[] #List to store price of the property
    areas=[] #List to store area of the property
    locations=[]
    beds=[]
    baths=[]
    dates=[]
    html_text_graana = requests.get(url).text
    soup_graana = BeautifulSoup(html_text_graana, 'lxml')
    try:
        try:
            for cost in soup_graana.find_all('div', class_='col-10'):
                price=cost.find('strong', class_='text-ellipsis')
                prices.append(price.text)
        except Exception as e:
            prices.append('None')
        for dif in   soup_graana.find_all('div', class_="style_details__fUTrB col-7 no-padding"):
            try:
                type_=dif.find('h2', class_="clickable space-0 text-ellipsis")
                types.append(type_.text)
            except Exception as e:
                types.append('None')
            try:    
                loc=dif.find('p', class_="clickable space-0 text-ellipsis greenHover")
                if len(loc.text.split(','))>2:
                    #locations.append(loc.text)
                    locations.append(loc.text.split(',')[1])
                else:
                    locations.append(loc.text.split(',')[0])
            except Exception as e:
                locations.append('None')
#             try:
#                 bedbath = dif.find('p', class_="gray small").text.split(sep="|")
#                 beds.append(bedbath[0])
#                 baths.append(bedbath[1])
#             except Exception as e:
#                 beds.append('None')
#                 baths.append('None')
            bedbath = dif.find('p', class_="gray small").text.split(sep='|')
    
            try:
                beds.append(int(bedbath[0]))
            except Exception as e:
                beds.append('')

            try:
                baths.append(int(bedbath[1]))
            except Exception as e:
                baths.append('')

            try:
                posted=dif.find('p', class_='small text-ellipsis no-margin style_addedBy__2cBwa')
                date=posted.find('strong')
                dates.append(date.text)
            except Exception as e:
                dates.append('None')

    except Exception as e:
        print(e)

    
    for dia in soup_graana.find_all('div',class_='clickable col-12 col-xs-12 hidden-xs'):
        try:
            area=dia.find('strong')
            areas.append(area.text)
        except Exception as e:
            areas.append('')
        
    #return prices,areas,beds,baths,dates,locations,types 
    a={'Price':prices,'Area':areas,'Beds':beds,'Baths':baths,'Posted':dates,'Location':locations,'Property Type':types}
    
    df=pd.DataFrame.from_dict(a, orient='index')
    
    df.transpose().to_csv (r'Graana'+ time.strftime("%d%m%Y")+'.csv', index = True, header=True)
    
    return df.transpose()
```

```python
from datetime import datetime, timedelta



def date_convert(date):
    try:
        if "hours" in date:
            return datetime.now().strftime('%Y-%m-%d')
        elif "days"  in date:
            split = date.split(sep=' ')
            N = int(split[1])
            date_N_days_ago = datetime.now() - timedelta(days=N)
            return date_N_days_ago.strftime('%Y-%m-%d')
        elif "minutes" in date:
            split = date.split(sep=' ')
            N = int(split[1])
            date_N_days_ago = datetime.now() - timedelta(days=N)
            return date_N_days_ago.strftime('%Y-%m-%d')
        elif "weeks" in date:
            splitw = date.split(sep=' ')
            N_W = int(splitw[1]) * 7
            date_N_week_ago = datetime.now() - timedelta(days=N_W)
            return date_N_week_ago.strftime('%Y-%m-%d')
        elif "months" in date:
            splitm = date.split(sep=' ')
            N_M = int(splitm[1]) * 30
            date_N_month_ago = datetime.now() - timedelta(days=N_M)
            return date_N_month_ago.strftime('%Y-%m-%d')
        elif "hour" in date:
            return datetime.now().strftime('%Y-%m-%d')
        elif "minute" in date:
            split = date.split(sep=' ')
            N = int(split[1])
            date_N_days_ago = datetime.now() - timedelta(days=N)
            return date_N_days_ago.strftime('%Y-%m-%d')
        elif "day"  in date:
            split = date.split(sep=' ')
            N = int(split[1])
            date_N_days_ago = datetime.now() - timedelta(days=N)
            return date_N_days_ago.strftime('%Y-%m-%d')
        elif "week" in date:
            splitw = date.split(sep=' ')
            N_W = int(splitw[1]) * 7
            date_N_week_ago = datetime.now() - timedelta(days=N_W)
            return date_N_week_ago.strftime('%Y-%m-%d')
        elif "month" in date:
            splitm = date.split(sep=' ')
            N_M = int(splitm[1]) * 30
            date_N_month_ago = datetime.now() - timedelta(days=N_M)
            return date_N_month_ago.strftime('%Y-%m-%d')
    except Exception as e:
        return (datetime.now() - timedelta(days=10000)).strftime('%Y-%m-%d')
# N = 2

# date_N_days_ago = datetime.now() - timedelta(days=N)

# print(datetime.now())
# print(date_N_days_ago)
```

```python
def zameen(url):
    types=[] #List of type of property
    prices=[] #List to store price of the property
    areas=[] #List to store area of the property
    locations=[]
    beds=[]
    baths=[]
    dates=[]
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'lxml')
    try:
        for div in soup.find_all('li', class_='ef447dde'):
            try:
                price = div.find('span', class_='f343d9ce')
                prices.append(price.text)
            except Exception as e:
                prices.append('NaN')
            try:    
                location=div.find('div', class_='_162e6469')
                locations.append(location.text.split(',')[0])
            except Exception as e:
                locations.append('None')
            try:    
                date=div.find('span', {'aria-label':'Listing creation date'})
                days = date_convert(date.text)
                dates.append(days)
            except Exception as e:
                dates.append('None')
            try:
                type_=div.find('h2', class_='c0df3811')
                types.append(type_.text)
            except Exception as e:
                types.append('None')
    except Exception as e:
        print(e)
    
    
    try:
        for specs in soup.find_all('div', class_='_22b2f6ed'):
                try:
                        bed=specs.find('span', {'aria-label':'Beds'})
                        beds.append(bed.text)
                except Exception as e:
                        beds.append("1")
                try:
                        bath=specs.find('span', {'aria-label':'Baths'})
                        baths.append(bath.text)
                except Exception as e:
                        baths.append("1")
                try:
                        area=specs.find('span', {'aria-label':'Area'})
                        areas.append(area.text)
                except Exception as e:
                        areas.append("None")
    except Exception as e:
        print(e)
        
    #return prices,areas,beds,baths,dates,locations,types
    a={'Price':prices,'Area':areas,'Beds':beds,'Baths':baths,'Posted':dates,'Location':locations,'Property Type':types}
    df=pd.DataFrame.from_dict(a, orient='index')
    df.transpose().to_csv (r'Zameen'+ time.strftime("%d%m%Y")+'.csv', index = True, header=True)
    return df.transpose()
```

```python

#df_g_3_5_1 = list_to_df(pricesg_3_5_1,areasg_3_5_1,bedsg_3_5_1,bathsg_3_5_1,datesg_3_5_1,locationsg_3_5_1,typeg_3_5_1)
def graana_wrang(df):

    #Remove Rows containing any None Values
    df = df.replace(r'^\s*$', np.nan, regex=True)
    df = df.mask(df.astype(object).eq('None')).dropna()

    #extract only numeric value from price
    df['Price'] = df['Price'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)")

    df = df.copy()
    df_home=df.copy()
    #seperate sqft values from df
    df=df[df['Area'].str.contains('Sqft|Sq ft|sqft|sq ft', na=False)]
    #extract number valuse from area
    df['Area'] = df['Area'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)")
    #convert area integer into integer
    df['Area'] = pd.to_numeric(df['Area'])
    #convert area into marla
    df['Area'] = (df['Area'] / 225).round(1)
    #change name to apartment
    df['Property Type'] = 'Apartment'

#For house
    df_home=df_home[df_home['Area'].str.contains('House|house|Home|home', na=False)]
    #extract number valuse from area
    df_home['Area'] = df_home['Area'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)")
    df_home['Area'] = pd.to_numeric(df_home['Area'])
    #df['Area'] = (df['Area'] / 225).round(1)
    #change name to apartment
    df_home['Property Type'] = 'House'
#Correcting dates
#Coverting Bedgs and Baths to integer
   # df_home['Beds']=pd.to_numeric(df_home['Beds'])
    #df_home['Beds'] = pd.to_numeric(df_home['Beds'], downcast='float')
    #df_home['Baths'] = pd.to_numeric(df_home['Baths'])

    
    
    result = pd.concat([df,df_home],ignore_index=True)
    result['Beds']=pd.to_numeric(result['Beds'],downcast='integer')
    result['Baths']=pd.to_numeric(result['Baths'],downcast='integer')
    result['Posted']=result['Posted'].str.replace(',',' ')
    result.rename(columns = {'Posted': 'Date'}, inplace = True)
    result[['mm', 'dd', 'year']] = pd.DataFrame([ x.split() for x in result['Date'].tolist() ])
   # result[['mm','dd']] = result.Date.str.split(' ', 1, expand=True)
   # result[['day','year']] = result.dd.str.split(' ', 1, expand=True)
    #clean_df['year']=clean_df['year'].str.strip()
    result['Posted'] = pd.to_datetime(result['dd'] + result['mm'] + result['year'],format='%d%B%Y')
    result.drop(['Date', 'mm','dd','year'], axis='columns', inplace=True)
    result = result.sort_values(by="Posted", ascending=False)
    # Coverting Price Column Type to Numeric
    result['Price'] = pd.to_numeric(result['Price'])
    return result.reset_index(drop=True)
```

```python
def zameen_wrang(df):
    #Remove Rows containing any None Values
    df = df.replace(r'^\s*$', np.nan, regex=True)
    df = df.mask(df.astype(object).eq('None')).dropna()
    #extract only numeric value from price
    df['Price'] = df['Price'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)")
    #extract number valuse from area
    df['Area'] = df['Area'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)")
    
    # Separate Houses
    df_home=df.copy()
    df_home=df_home[df_home['Property Type'].str.contains('House|house|Home|home', na=False)]
    df_home['Property Type'] = 'House'
    df=df[df['Property Type'].str.contains('Apartment|Flat|apartment|flat', na=False)]
    df['Property Type'] = 'Apartment'
    result = pd.concat([df,df_home],ignore_index=True)
    result['Price'] = pd.to_numeric(result['Price'])
    result['Area'] = pd.to_numeric(result['Area'])
    result['Beds']=pd.to_numeric(result['Beds'],downcast='integer')
    result['Baths']=pd.to_numeric(result['Baths'],downcast='integer')
    result['Posted'] = pd.to_datetime(result['Posted'])
    result = result.sort_values(by="Posted", ascending=False)
    return result.reset_index(drop=True)
```

## Combine Both Graana and Zameen DataFrames

```python
def fetch_record(urlg,urlz):

    #Call graana Function and return all lists.
    df_g = graana(urlg)

    #clean the dataframe
    clean_g = graana_wrang(df_g)
    clean_g

    #call zameen function
    df_z = zameen(urlz)
    clean_z=zameen_wrang(df_z)

    combined_df = pd.concat([clean_g,clean_z], ignore_index=True)
    return combined_df
```

## Fetch Data

```python
urlg='https://www.graana.com/residential/for_sale/islamabad/all/1?price_min=500000&price_max=7000000&page=1'
urlz='https://www.zameen.com/Homes/Islamabad-3-1.html?price_max=7000000'

combine_result = fetch_record(urlg,urlz)
combine_result = combine_result.sort_values(by="Posted", ascending=False)
combine_result.reset_index(drop=True,inplace=True)
combine_result
```

## Add Latitude and Longitude

```python
geolocator = Nominatim(user_agent='my_app')

for i in range(len(combine_result)):
    try:
        location = geolocator.geocode(combine_result.loc[i,'Location'] + ' ' + 'Islamabad')
        combine_result.loc[i,'Latitude'] = location.latitude   
        combine_result.loc[i,'Longitude'] = location.longitude
    except Exception as e:
        combine_result.loc[i,'Latitude'] =00.0000000
        combine_result.loc[i,'Longitude'] =00.0000000
combine_result
```
